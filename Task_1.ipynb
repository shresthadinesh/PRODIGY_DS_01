## Top ten countries of total population in year 2022 and 2016

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar Plot for year 2022

plt.figure(figsize = (15,6))
sns.barplot(x = '2022', y = 'Country Name', data = total_top_fifteen_countries,hue="Country Name", palette="coolwarm", legend=False)
plt.title("Top 10 countries of total population in year 2022", fontsize = 15)
plt.xlabel("Total Population", fontsize = 12)
plt.ylabel("Country Name", fontsize = 12)
plt.show
```

```python
# Bar plot for year 2001

plt.figure(figsize = (15,6))
sns.barplot(x = '2001', y = 'Country Name', data = total_top_fifteen_countries, hue = 'Country Name', palette = 'coolwarm', legend = False)
plt.title("Top 15 Countries of total population in year 2001", fontsize = 15)
plt.xlabel("Total Population", fontsize = 12)
plt.ylabel("Country Name", fontsize = 12)
plt.show()
```

```python
# Select top 15 countries based on 2022 population
total_population_sorted = df[df['Series Code'] == 'SP.POP.TOTL'].sort_values(by="2022", ascending=False)
total_top_fifteen_countries = total_population_sorted.head(15)

# Convert the data to long format for easier plotting
df_long = total_top_fifteen_countries.melt(id_vars=["Country Name"], value_vars=["2001", "2022"], 
                                           var_name="Year", value_name="Population")

# Plot the bar chart with hue for year comparison
plt.figure(figsize=(15, 6))
sns.barplot(x="Population", y="Country Name", hue="Year", data=df_long, palette="muted")

# Titles and labels
plt.title("Top 15 Countries by Total Population (2001 vs 2022)", fontsize=12)
plt.xlabel("Total Population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.legend(title="Year")
plt.show()
```

# Stacked Bar Plot


```python
# merge male and female population on "Country Name"
merged_data = pd.merge(total_male_population, total_female_population, on = 'Country Name', suffixes = ('_male','_female'))
merged_data.columns
```

```python
# Calculate total population for each country (male + female)
merged_data['Total Population'] = merged_data['2022_male'] + merged_data['2022_female']
```

```python
merged_data.columns
```

```python
merged_data.head()
```

```python
# sort data based on total population 
sort_data = merged_data.sort_values(by = 'Total Population', ascending = False)
sort_data[['Country Name', 'Total Population']].head(10)
                                        
```

```python
Top_ten_countries = sort_data.head(10)
```

```python
# Create the stacked bar plot
plt.figure(figsize=(12, 6))

sns.barplot(x="Country Name", y="2022_female", data=Top_ten_countries, color="red", label="Female Population")
sns.barplot(x="Country Name", y="2022_male", data=Top_ten_countries, bottom=Top_ten_countries["2022_female"], color="green", label="Male Population")
plt.title("Top 10 Countries with Male and Female Populations (2022)", fontsize = 15)
plt.xlabel("Country Name", fontsize = 12)
plt.ylabel("Total Population", fontsize = 12)
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
